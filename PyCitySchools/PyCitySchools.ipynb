```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
```

```python
# Local Government Area Summary

df
```

```python
# df of individual schools

grouped_schools = df.groupby(["school_name"]).mean(['maths_score'])
grouped_schools
```

```python
grouped_schools.shape[0]
```

```python
# total budget
total_budget = grouped_schools["budget"].sum()
total_budget
```

```python
# total number of schools
num_schools = grouped_schools.shape[0]
num_schools
```

```python
# total schools (unique)
schools_unique_list = df["school_name"].unique()
total_schools = len(schools_unique_list)
total_schools
schools_unique_list
```

```python
# total number of students
total_students = grouped_schools["size"].sum()
total_students
```

```python

# calculate mean math score
mean_math_score = df['maths_score'].mean()
rounded_mean_math_score = mean_math_score.round(2)

# calculate mean math score
mean_read_score = df['reading_score'].mean()
rounded_mean_read_score = mean_read_score.round(2)
#print (f"mean_math_score: {mean_math_score}")

#data = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
#df = pd.DataFrame(data, index=['rank1','rank2','rank3','rank4'])
```

```python
num_students_math_pass_df = df.loc[df['maths_score'] >= 50]
num_students_math_pass_df
```

```python
num_students_read_pass_df = df.loc[df['reading_score'] >= 50]
num_students_read_pass_df
```

```python
# calculate % pass rates - math

perc_math_pass = num_students_math_pass_df.shape[0]/total_students*100
perc_math_pass
```

```python
# calculate % pass rates - read

perc_read_pass = num_students_read_pass_df.shape[0]/total_students*100
perc_read_pass
```

```python
# calculate number of students who passed both math and reading
num_students_math_read_pass_df = df.loc[(df['maths_score'] >= 50) & (df['reading_score'] >= 50)]
num_students_math_read_pass_df
```

```python
# calculate % that passed both math and reading
perc_math_read_pass = num_students_math_read_pass_df.shape[0]/total_students*100
perc_math_read_pass
```

```python
# Summary Report
data = {'Total_Schools':[total_schools],
        'Total_Students':[total_students],
        'Total_Budget':[total_budget],
        'Avg_Math_Score':[rounded_mean_math_score],
        'Avg_Reading_Score':[rounded_mean_read_score],
        '% Pass Math':[perc_math_pass],
        '% Pass Reading':[perc_read_pass],
        '% Pass Reading&Math':[perc_math_read_pass]
       }
result_df = pd.DataFrame(data, index=['Results'])

result_df
```
